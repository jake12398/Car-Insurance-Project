Customer Table

```python
from faker import Faker
import random
import sqlite3
import numpy as np

from datetime import date
from dateutil.relativedelta import relativedelta
from datetime import date, timedelta



conn = sqlite3.connect("Insurance project.db")
cursor = conn.cursor()
fake = Faker()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Customers(
    CustomerID INTEGER PRIMARY KEY AUTOINCREMENT,
    Age INTEGER,
    Name TEXT,
    Location TEXT,
    Gender TEXT,
    DatePassedDrivingTest Date
)
""")

min_age = 18
max_age = 100
a, b = 1, 5

def generate_person():
    gender = random.choice(["M", "F"])

    if gender == "M":
        first = fake.first_name_male()
    else:
        first = fake.first_name_female()

    last = fake.last_name()

    # generate ONE age only
    u = np.random.beta(a, b)
    age = int(min_age + (max_age - min_age) * u)

    return f"{first} {last}", gender, age

data = []

for _ in range(10000):
    name, gender, age = generate_person()
    location=fake.city()
    end_date = date(2026, 4, 30)
    start_date = end_date - relativedelta(years=age-17)
    delta = end_date - start_date

    random_days = random.randint(0, delta.days)
    Driving_test_date = start_date + timedelta(days=random_days)
    data.append((age, name, location, gender,Driving_test_date))

cursor.executemany("""
INSERT INTO Customers (Age, Name, Location, Gender, DatePassedDrivingTest)
VALUES (?, ?, ?, ?, ?)
""", data)

conn.commit()
conn.close()
```

Vehicle Table

```python
import random
import numpy as np
import sqlite3
import math
conn = sqlite3.connect("Insurance project.db")
cursor = conn.cursor()

# ✅ IMPORTANT: wipe old broken schema
cursor.execute("DROP TABLE IF EXISTS Vehicle")

cursor.execute("""
CREATE TABLE Vehicle(
    VehicleID INTEGER PRIMARY KEY AUTOINCREMENT,
    Brand TEXT,
    Age INTEGER,
    Mileage INTEGER,
    Value INTEGER
)
""")
Vehicle_brand=["Ford", "Toyota", "Volkswagen", "Honda", "Nissan", "Hyundai", "Kia", "Chevrolet", "BMW", "Mercedes-Benz", "Audi", "Peugeot", "Renault", "Skoda", "SEAT", "Mazda", "Subaru", "Suzuki", "Tesla", "Volvo", "Jaguar", "Land Rover", "Porsche", "Lexus", "Ferrari", "Lamborghini"]
low_end=["Ford", "Toyota", "Volkswagen", "Hyundai", "Kia", "Peugeot", "Renault", "Skoda", "SEAT","Suzuki","Lexus"]
medium_end=["Honda", "Nissan","Mazda", "Subaru","Volvo", "Land Rover","Tesla"]
high_end=["Chevrolet", "BMW", "Mercedes-Benz", "Audi","Jaguar", "Porsche"]
luxery_end=["Ferrari", "Lamborghini"]

data = []

for _ in range(10000):
    age=random.randint(1,15)
    mpy=np.random.normal(8500, 3000)
    mpy=int(round(mpy))
    while mpy < 0:
        mpy = np.random.normal(8500, 3000)
    mpy = np.clip(mpy, 3000, 15000)
    mileage=age*mpy
    mileage=int(round(mileage))
    
    car=random.choice(Vehicle_brand)
    if car in low_end:
        price=random.randint(10000, 25000)*math.exp(-(5*10**-5)*(mileage/10000))
    elif car in medium_end:
        price=random.randint(20000,45000)*math.exp(-(2*10**-5)*(mileage/10000))
    elif car in high_end:
        price=random.randint(40000, 100000)*math.exp(-(1*10**-5)*(mileage/10000))
    elif car in luxery_end:
        price=random.randint(100000, 200000)*math.exp(-(5*10**-6)*(mileage/10000))
    data.append((car, age, int(mileage), int(price)))
   
cursor.executemany("""
INSERT INTO Vehicle (Brand, Age, Mileage, Value)
VALUES (?, ?, ?, ?)
""", data)

conn.commit()
conn.close()
```

Policy Table

```python
import sqlite3
import random
import math
from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta


conn = sqlite3.connect("Insurance project.db")
cur = conn.cursor()

##Pairing cars to customers(age effects the car they are paired with)
cur.execute("""
WITH ranked_customers AS (
    SELECT
        CustomerID,
        Age,
        DatePassedDrivingTest,
        ROW_NUMBER() OVER (ORDER BY Age + (RANDOM() % 5)) AS rn
    FROM Customers
),
ranked_vehicles AS (
    SELECT
        VehicleID,
        Value,
        ROW_NUMBER() OVER (ORDER BY Value + (RANDOM() % 50)) AS rn
    FROM Vehicle
)
SELECT
    c.CustomerID,
    c.Age,
    c.DatePassedDrivingTest,
    v.VehicleID,
    v.Value
FROM ranked_customers c
JOIN ranked_vehicles v
ON c.rn = v.rn;
""")

pairs = cur.fetchall()

#Create policy table
cur.execute("""
CREATE TABLE IF NOT EXISTS Policy (
    PolicyID INTEGER PRIMARY KEY AUTOINCREMENT,
    CustomerID INTEGER,
    VehicleID INTEGER,
    StartDate Date,
    EndDate Date,
    CoverageType TEXT,
    Excess REAL,
    Premium REAL
)
""")

conn.commit()
#Coverage Type
Coverage_type=["Third party", "Third party, fire and theft", "Fully Comprehensive"]

#Excess calculation
def excess(age, value, coverage):

    # base excess 
    base = 250 + (value * 0.01) 

    # age effect
    age_factor = max(0, (35 - age)) ** 1.2 * 10
    
    coverage_modifier = {
        "Third party": 1.4,
        "Third party, fire and theft": 1.1,
        "Fully Comprehensive": 0.7
    }

    modifier = coverage_modifier.get(coverage, 1.0)

    # small noise
    noise = random.uniform(-100, 100)

    result = (base + age_factor + noise) * modifier

    #realistic insurance bounds
    return round(max(150, min(result, 1500)), 2)


def Policy_startdate(DatePassedDrivingTest):
    dt = datetime.strptime(DatePassedDrivingTest, "%Y-%m-%d")

    delay_days = random.randint(0, 30)

    return (dt + timedelta(days=delay_days)).date()
def years_since_test(driving_test_date):

    dt = datetime.strptime(driving_test_date, "%Y-%m-%d")
    today = date(2026, 4, 30)

    diff = relativedelta(today, dt)

    return diff.years

#Premium calculation
def premium(age, value, coverage, excess_value, years):
    #Base premium
    base=200
    
    #age risk
    age_risk=3000/(age-17+1)
    
    #Driving experience
    experience_discount=400/(years+1)
    
    #Value
    value_risk=math.log(value+1)*25
    
    #Coverage
    coverage_modifier = {
        "Third party": 1.0,
        "Third party, fire and theft": 1.25,
        "Fully Comprehensive": 1.6
    }.get(coverage, 1.0)
    
    #excess discount
    excess_discount = excess_value * 0.3
    
    #Calculation
    raw_premium = base + age_risk + value_risk - experience_discount

    premium_value = (raw_premium * coverage_modifier) - excess_discount
    return round(max(200, min(premium_value, 8000)), 2)
    
#Insert restults
cur.execute("DELETE FROM Policy")

for customer_id, age, DatePassedDrivingTest, vehicle_id, value in pairs:
    
    #CoverageType
    Coverage=random.choice(Coverage_type)
    
    #Start date
    start_date=Policy_startdate(DatePassedDrivingTest)
    
    #Enddate
    num= random.randint(1, 5)
    End_date=start_date+relativedelta(years=num)
    
    #Excess
    excess_value = excess(age, value, Coverage)
    
    #Years since passing driving test
    years = years_since_test(DatePassedDrivingTest)
    
    cur.execute("""
        INSERT INTO Policy (CustomerID, VehicleID, StartDate, EndDate, CoverageType, Excess, Premium)
        VALUES (?, ?, ?, ?, ?, ?, ?)
    """, (
        customer_id,
        vehicle_id,
        start_date,
        End_date,
        Coverage,
        excess_value,
        premium(age, value, Coverage, excess_value, years)
    ))

conn.commit()
conn.close()
```

Claims Table

```python
import sqlite3
import random
from datetime import datetime, timedelta

conn = sqlite3.connect("Insurance project.db")
cursor = conn.cursor()

# Create Claims table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Claims(
    ClaimsID INTEGER PRIMARY KEY AUTOINCREMENT,
    PolicyID INTEGER,
    ClaimDate DATE,
    ClaimType TEXT,
    ClaimStatus TEXT,
    Amount REAL,
    FOREIGN KEY (PolicyID) REFERENCES Policy(PolicyID)
)
""")

conn.commit()

# Get policy data
cursor.execute("""
SELECT p.PolicyID, p.StartDate, p.EndDate, p.CoverageType, p.Excess, v.Value
FROM Policy p
JOIN Vehicle v ON p.VehicleID = v.VehicleID
""")

policies = cursor.fetchall()

# Claim types
claim_types = [
    "Glass Damage", "Vandalism", "Accident(minor)",
    "Accident(major)", "Weather", "Theft", "Fire"
]

# Function to generate claim date
def claim_date(start, end):
    dt1 = datetime.strptime(start, "%Y-%m-%d")
    dt2 = datetime.strptime(end, "%Y-%m-%d")
    delta = dt2 - dt1
    random_day = random.randint(0, delta.days)
    return (dt1 + timedelta(days=random_day)).date()
#Amount calculations
def Amount(value, excess,damage):
    
    claim_modifier = {
         "Glass Damage":0.1, "Vandalism":0.3, "Accident(minor)":0.4,
    "Accident(major)":0.8, "Weather":0.5, "Theft":1.0, "Fire":0.9
    }

    modifier = claim_modifier.get(damage, 1.0)


    result = (value*modifier) -excess

    #realistic insurance bounds
    return round(max(0,result), 2)


data = []

# Generate claims
for policy_id, start, end, coverage,excess,value in policies:

    # number of claims per policy (0–3 for realism)
    num_claims = random.randint(0, 3)

    for _ in range(num_claims):

        damage = random.choice(claim_types)
        date_of_claim = claim_date(start, end)

        # Claim status logic
        if coverage == "Third party":
            if damage in ["Accident(minor)", "Accident(major)"]:
                claim_status = "Approved"
            else:
                claim_status = "Denied"

        elif coverage == "Third party, fire and theft":
            if damage in ["Accident(minor)", "Accident(major)", "Theft", "Fire"]:
                claim_status = "Approved"
            else:
                claim_status = "Denied"

        elif coverage == "Fully Comprehensive":
            claim_status = "Approved"

        else:
            claim_status = "Pending"
        amount=Amount(value,excess,damage)

        data.append((policy_id, date_of_claim, damage, claim_status,amount))

# Insert into Claims table
cursor.executemany("""
INSERT INTO Claims (PolicyID, ClaimDate, ClaimType, ClaimStatus, Amount)
VALUES (?, ?, ?, ?, ?)
""", data,)

conn.commit()
conn.close()
```
