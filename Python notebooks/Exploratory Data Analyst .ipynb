```python
import sqlite3
import matplotlib.pyplot as plt
import random

conn = sqlite3.connect("Insurance project2.db")
cursor = conn.cursor()
query=(""" SELECT 
	c.Location,
	sum(Cl.Amount) AS Totalclaim
FROM Claims cl
JOIN policy p
    ON cL.pOLICYid = p.POLICYID
JOIN customers c
	ON p.CustomerID = c.CustomerID
GROUP BY c.Location
ORDER BY Totalclaim DEsc Limit 20;""")
cursor.execute(query)
results = cursor.fetchall()

conn.close()
Location=[column[0]for column in results]
Claim=[column[1]for column in results]
plt.figure(figsize=(12,6))

plt.bar(Location, Claim)

plt.xlabel("Location")
plt.ylabel("Total Claim Amount")
plt.title("Top 10 Locations by Highest Claim Amount")

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
conn = sqlite3.connect("Insurance project2.db")
cursor = conn.cursor()
query=(""" SELECT 
	c.Age,
	sum(Cl.Amount) AS TotalClaims
FROM Claims cl
JOIN policy p
    ON cL.pOLICYid = p.POLICYID
JOIN customers c
	ON p.CustomerID = c.CustomerID
GROUP BY Age;""")
cursor.execute(query)
results = cursor.fetchall()

conn.close()
Age=[column[0]for column in results]
Claim=[column[1]for column in results]
plt.figure(figsize=(12,6))

plt.bar(Age, Claim)
import matplotlib.ticker as ticker

ax = plt.gca()
ax.yaxis.set_major_formatter(
    ticker.StrMethodFormatter('{x:,.0f}')
)

plt.xlabel("Age")
plt.ylabel("Total Claim Amount")
plt.title("Age against Claims Amount")

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
conn = sqlite3.connect("Insurance project2.db")
cursor = conn.cursor()
query=(""" SELECT 
    ClaimType,
    SUM(Amount) AS TotalClaims
FROM Claims
GROUP BY ClaimType;""")
cursor.execute(query)
results = cursor.fetchall()

conn.close()
Claimamount=[column[1]for column in results]
ClaimType=[column[0]for column in results]
plt.figure(figsize=(12,6))

plt.bar(ClaimType,Claimamount)

import matplotlib.ticker as ticker

ax = plt.gca()
ax.yaxis.set_major_formatter(
    ticker.StrMethodFormatter('{x:,.0f}')
)

plt.xlabel("Claim Type")
plt.ylabel("Total Claim Amount")
plt.title("Claim Type against Claims Amount")

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```
